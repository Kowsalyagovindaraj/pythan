```python
import gradio as gr
import mysql.connector

# Establish a connection to the MySQL database
#mydb = mysql.connector.connect(
 # host="localhost",
  #user="nethaji",
  #password="Sample@123",
  #database="rebin"
#)

# Create a cursor
mycursor = mydb.cursor()

# Define a function to insert data into the table
def insert_data(id, name, address):
    try:
        sql = "insert into kowsi01.guvi(id, name, address) VALUES (%s, %s, %s)"
        values = (id, name, address)
        mycursor.execute(sql, values)
        mydb.commit()
        return "Data inserted successfully!"
    except Exception as e:
        return f"Error: {str(e)}"

# Gradio interface
with gr.Interface(
    fn=insert_data,
    inputs=[gr.Number(label="ID"), gr.Textbox(label="Name"), gr.Textbox(label="Address")],
    outputs="text",
    title="insert Data into kowsi01 Table"
) as demo:
    demo.launch()
```

```python
#complexquery 
mycursor.execute("SELECT CASE \
        WHEN bmi < 18.5 THEN 'Underweight' \
        WHEN bmi BETWEEN 18.5 AND 24.9 THEN 'Normal' \
        WHEN bmi BETWEEN 25 AND 29.9 THEN 'Overweight' \
        ELSE 'Obese' \
      END \
     AS bmi_category, \
   COUNT(*) AS patient_count \
FROM guvi.diabetes \
GROUP BY bmi_category")
out = mycursor.fetchall() 
from tabulate import tabulate
print(tabulate(out, headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
mycursor.execute("""
    SELECT 
        bmi,  -- This will select all columns from the diabetes table
        CASE 
            WHEN bmi < 18.5 THEN 'Underweight'
            WHEN bmi BETWEEN 18.5 AND 24.9 THEN 'Normal'
            WHEN bmi BETWEEN 25 AND 29.9 THEN 'Overweight'
            ELSE 'Obese'
        END AS bmi_category
    FROM guvi.diabetes
    ORDER BY bmi
""")
out = mycursor.fetchall() 
from tabulate import tabulate
print(tabulate(out, headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```
