```python
#Box plot
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate sample data

np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=60, scale=15, size=100)
data3 = np.random.normal(loc=70, scale=20, size=100)

# Combine data into a list
data = [data1, data2, data3]

# Labels for the datasets
labels = ['Group 1', 'Group 2', 'Group 3']

# Matplotlib Box Plot
plt.boxplot(data, labels=labels, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title('Box Plot with Multiple Data')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.show()

sns.boxplot(data=data, palette='pastel')
plt.xticks(ticks=[0, 1, 2], labels=labels)
plt.title('Seaborn Box Plot with Multiple Data')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Sample data with country code
data = {'country_code': ['US', 'IN', 'US', 'FR', 'IN', 'US'],
        'restaurant_name': ['Pizza Hut', 'Taj Mahal', 'Burger King', 'La Maison', 'Dosa Place', 'KFC']}
df = pd.DataFrame(data)
# Count restaurants by country code
restaurant_counts = df['country_code'].value_counts()
# Create a bar plot
plt.figure(figsize=(8, 6))
restaurant_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Restaurants by Country Code')
plt.xlabel('Country Code')
plt.ylabel('Restaurant Count')
plt.xticks(rotation=0)  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

```python

# prompt: write a plot  that represents total number of restaurants for each cuisines
import matplotlib.pyplot as plt
import pandas as pd
# Load the dataset
data = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/refs/heads/main/zomato/zomato.csv")
# Count the number of restaurants for each cuisine
cuisine_counts = data['Cuisines'].value_counts()
# Create the bar plot
plt.figure(figsize=(12, 6))
plt.bar(cuisine_counts.index[:20], cuisine_counts.values[:20], color='skyblue')  # Plotting only the top 20 cuisines for better visualization
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.title('Total Number of Restaurants for Each Cuisine (Top 20)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: draw a sunburst chart using the above df data
import pandas as pd
import plotly.express as px
# Load the dataset (assuming 'df' is already loaded as shown in your code)
df = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/COMMONWEALTH%20GAMES%201930-2018%20%20-%20MEDAL%20TALLY.csv")
# Prepare the data for the sunburst chart
# Assuming you want to visualize medals by Year, Country, and Total medals
# Replace 'Total' with the actual column name if different
df['Total'] = pd.to_numeric(df['Total'], errors='coerce')  # Convert to numeric, handle errors
fig = px.sunburst(df, path=['Year', 'Country'], values='Total',
                  title="Commonwealth Games Medals by Year and Country")
fig.show()
```

```python
# prompt: plot the location using latitude and longitude with the restaurent id from df
import plotly.express as px
# Assuming 'hi' DataFrame is already loaded and has columns 'Restaurant ID', 'Latitude', and 'Longitude'
# Replace column names if they are different in your dataframe.
fig = px.scatter_mapbox(df, lat="Latitude", lon="Longitude", hover_name="Restaurant ID", hover_data=["Restaurant ID"],
                        color_discrete_sequence=["fuchsia"], zoom=1, height=300)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()



#"open-street-map": Standard street map.
#"carto-positron": Light-themed map.
#"carto-darkmatter": Dark-themed map.
```

```python
# prompt: draw a px.scatter animation using the above medal data with country year and total medals avoid nan values
import plotly.express as px
import pandas as pd
medal = pd.read_csv("https://raw.githubusercontent.com/nethajinirmal13/Training-datasets/main/COMMONWEALTH%20GAMES%201930-2018%20%20-%20MEDAL%20TALLY.csv")
# Convert 'Total' column to numeric, handling errors
medal['Total'] = pd.to_numeric(medal['Total'], errors='coerce')
# Drop rows with NaN values in 'Total'
medal = medal.dropna(subset=['Total'])
# Create the scatter animation
fig = px.scatter(medal, x="Country", y="Total", animation_frame="Year", 
                 animation_group="Country", size="Total", color="Country", 
                 hover_name="Country", title="Medal Count Animation")
fig.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
# Sample data 
data = {'country_code': [1, 2, 1, 3, 2, 1],
        'restaurant_name': ['A', 'B', 'C', 'D', 'E', 'F'],
        'latitude': [34.0522, 19.0760, 37.7749, 48.8566, -33.8688, 51.5074],
        'longitude': [-118.2437, 72.8777, -122.4194, 2.3522, 151.2093, 0.1278]}
df = pd.DataFrame(data)
# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
# Plot 1: 
axes[0, 0].scatter(df['country_code'], df['latitude'])
axes[0, 0].set_title('Country Code vs. Latitude')
axes[0, 0].set_xlabel('Country Code')
axes[0, 0].set_ylabel('Latitude')
# Plot 2: Bar plot
restaurant_counts = df['country_code'].value_counts()
axes[0, 1].bar(restaurant_counts.index, restaurant_counts.values)
axes[0, 1].set_title('Restaurant Count by Country')
axes[0, 1].set_xlabel('Country Code')
axes[0, 1].set_ylabel('Restaurant Count')


axes[1,1].axis('off') # remove the axis from the subplot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the tips dataset
tips = sns.load_dataset("tips")
# Calculate the number of male and female smokers
smoker_counts = tips[tips['smoker'] == 'Yes']['sex'].value_counts()
# Create a bar plot
plt.figure(figsize=(6, 4))
plt.bar(smoker_counts.index, smoker_counts.values, color=['skyblue', 'lightcoral'])
plt.title('Number of Male and Female Smokers')
plt.xlabel('Sex')
plt.ylabel('Number of Smokers')
plt.show()
```
